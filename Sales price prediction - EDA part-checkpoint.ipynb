# SCATTER PLOT OF TV, RADIO AND NEWSPAPER 

```python
sn.scatterplot(x='TV',y='Sales',data=dataset)
plt.title("Understand the sales on TV adversiting")
plt.show()
```

Here we can not come to a conclusion that if we spending more in radio sales is also increasing maybe sales is increase maybe not or maybe down
